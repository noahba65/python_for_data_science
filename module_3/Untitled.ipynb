```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(211)
ax1.plot([1, 2, 3], [1, 2, 3])
ax2 = fig.add_subplot(2,1,2,
        facecolor='y')
ax2.plot([1, 2, 3], [3, 2, 1])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot([1, 2, 3], [1, 2, 3], 'r')
ax2 = fig.add_subplot(2,1,2,
        facecolor='b')
ax2.plot([1, 2, 3], [3, 2, 1], 'r')
```
